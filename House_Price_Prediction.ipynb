```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

data = {
    'Size': [500, 800, 1000, 1200, 1500, 1800, 2000, 2200, 2500],
    'Price': [100, 150, 200, 220, 300, 350, 400, 420, 500]
}

df = pd.DataFrame(data)

X = df[['Size']]
y = df['Price']

model = LinearRegression()
model.fit(X, y)

predicted_prices = model.predict(X)

plt.scatter(X, y, color='blue', label='Actual Prices', s=70)

plt.scatter(X, predicted_prices, color='red', label='Predicted Prices', s=70)

plt.plot(X, predicted_prices, color='green', linewidth=2)

for i in range(len(X)):
    plt.text(X.iloc[i, 0], y.iloc[i], f"{y.iloc[i]}", fontsize=8, color='blue')
    plt.text(X.iloc[i, 0], predicted_prices[i], f"{predicted_prices[i]:.1f}", fontsize=8, color='red')

plt.xlabel('House Size (sq ft)')
plt.ylabel('House Price ($1000)')
plt.title('Actual vs Predicted House Prices')
plt.legend()
plt.grid(True)
plt.show()

new_size = np.array([[1600]])
new_prediction = model.predict(new_size)
print(f"Predicted price for 1600 sq ft house: ${new_prediction[0]*1000:.2f}")
```
