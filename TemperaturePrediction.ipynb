```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt

Hours = np.array([[6], [9], [12], [15], [18]])
Tem = np.array([[18], [22], [30], [35], [28]])

model = LinearRegression()
model.fit(Hours, Tem)

hour = np.array([[14]])
predicted_temp = model.predict(hour)
print("Temperature at 14 hrs:", predicted_temp[0])

plt.figure(figsize=(8, 6))
plt.scatter(Hours, Tem, color='blue', label='Actual Temperature')
plt.scatter(hour, predicted_temp, color='red', marker='x', s=100, label=f'Predicted Temperature at {hour[0][0]} hrs')
plt.plot(Hours, model.predict(Hours), color='green', linestyle='--', label='Regression Line')
plt.xlabel('Hours')
plt.ylabel('Temperature')
plt.title('Temperature Prediction based on Hours')
plt.legend()
plt.grid(True)
plt.show()
```
